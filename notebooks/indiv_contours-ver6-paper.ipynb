```python
data = data_mcz40_sys3.copy()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))
fig.subplots_adjust(wspace=0.25)

# plot source waveform
lens_params = data["source_params"]
s_gw = get_gw(lens_params)
s_strain = np.abs(s_gw["strain"])
axes[0].plot(s_gw["f_array"], s_strain, label="lensed", c="magenta", ls="-")

# plot template waveforms
lens_params = data["source_params"]
RP_params = data["template_params"]
RP_params["omega_tilde"] = 0
RP_params["theta_tilde"] = 0
RP_params["gamma_P"] = 0
plot_template_waveform(
    RP_params,
    lens_params,
    get_updated_mismatch_results=True,
    axes=axes,
    label="unlensed",
    c="k",
    ls="--",
    phase_shift=2 * np.pi,
)

RP_params = data["template_params"]
RP_params["omega_tilde"] = data["stats"]["ep_min_omega_tilde"]
RP_params["theta_tilde"] = data["stats"]["ep_min_theta_tilde"]
RP_params["gamma_P"] = data["stats"]["ep_min_gammaP"]
plot_template_waveform(
    RP_params,
    lens_params,
    get_updated_mismatch_results=True,
    axes=axes,
    label="global",
    c="k",
    ls="-",
)

local_omega, local_theta = 1.5, 13.1
lens_params = data["source_params"]
RP_params = data["template_params"]
RP_params["omega_tilde"] = local_omega
RP_params["theta_tilde"] = local_theta
RP_params["gamma_P"] = data["gammaP_min_matrix"][
    np.where(
        (data["omega_matrix"] == local_omega) & (data["theta_matrix"] == local_theta)
    )
]
plot_template_waveform(
    RP_params,
    lens_params,
    get_updated_mismatch_results=True,
    axes=axes,
    label="local",
    c="blue",
    ls="-.",
    phase_shift=2 * np.pi,
)

customize_2x1_axes(axes)

# plt.savefig("../figures/sys3_waveforms_mcz40.pdf", bbox_inches="tight")
```

```python
data = data_mcz20_sys2.copy()


@interact(local_theta=(5, 8, 0.5))
def plot_waveform(local_theta):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))
    fig.subplots_adjust(wspace=0.25)

    # plot source waveform
    lens_params = data["source_params"]
    s_gw = get_gw(lens_params)
    s_strain = np.abs(s_gw["strain"])
    axes[0].plot(s_gw["f_array"], s_strain, label="source", c="magenta", ls="-")

    # plot template waveforms
    local_omega = 3
    lens_params = data["source_params"]
    RP_params = data["template_params"]
    RP_params["omega_tilde"] = local_omega
    RP_params["theta_tilde"] = local_theta
    RP_params["gamma_P"] = data["gammaP_min_matrix"][
        np.where(
            (data["omega_matrix"] == local_omega)
            & (data["theta_matrix"] == local_theta)
        )
    ]
    plot_template_waveform(
        RP_params,
        lens_params,
        get_updated_mismatch_results=True,
        axes=axes,
        label="local",
        c="k",
        ls="-",
        phase_shift=0,
    )

    customize_2x1_axes(axes)
    plt.show()
```
